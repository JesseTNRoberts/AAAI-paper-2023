```python
def load_pop_data(model_name):
    misra_data = pd.read_csv('rosch1975_alternate/'+model_name+'.csv')
    pop_data = pd.read_csv('rosch1975_alternate_popLM_50/'+model_name+'.csv')
    misra_data.head()
    misra_data['score (sum, mean, [list)'].head()
    pop_data['score (sum, mean, [list)'][0]

    mergedDF_50 = pd.merge(misra_data, pop_data, on=['item', 'category'], how='left')
    mergedDF_50 = mergedDF_50.drop(columns=['stimulus_y','rank_y', 'params_y', 'model_y'])
    mergedDF_50['score (sum, mean, [list)_x'] = mergedDF_50['score (sum, mean, [list)_x'].apply(ast.literal_eval)
    mergedDF_50['score (sum, mean, [list)_y'] = mergedDF_50['score (sum, mean, [list)_y'].apply(ast.literal_eval)

    mergedDF_50.rename(columns={'score (sum, mean, [list)_x': 'misra score (sum, mean, [list])'}, inplace=True)
    mergedDF_50.rename(columns={'score (sum, mean, [list)_y': 'popLM score (sum, mean, [list])'}, inplace=True)

    #print(mergedDF_50['popLM score (sum, mean, [list])'].head())
    #print(mergedDF_50['popLM score (sum, mean, [list])'][0])
    #print(len(mergedDF_50['popLM score (sum, mean, [list])'][0]))

    mergedDF_50['population data'] = [[t[2][0] for t in m] for m in mergedDF_50['popLM score (sum, mean, [list])']]
    mergedDF_50['misra_score'] = [t[2][0] for t in mergedDF_50['misra score (sum, mean, [list])']]    
    mergedDF_50['dropout_avg'] = mergedDF_50['population data'].apply(np.average)
    mergedDF_50['variance pop'] = mergedDF_50['population data'].apply(np.var)
    return mergedDF_50
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import linregress
import statsmodels

slope, intercept, r_value, p_value, std_err = linregress(gpt_2_toy['misra_score'], gpt_2_toy['dropout_avg'])

# Calculate the regression line
regression_line = [(slope * x) + intercept for x in gpt_2_toy['misra_score']]

# Create the scatter plot with regression line
fig = px.scatter(gpt_2_toy, x='misra_score', y='dropout_avg', title='Toy: Misra Vs. Dropout Average Linear Regression', trendline='ols')

# Add the regression line to the plot
fig.add_traces(go.Scatter(x=gpt_2_toy['misra_score'], y=regression_line, mode='lines', name='Regression Line'))

# Create summary statistics text
summary_text = f"Slope: {slope:.2f}<br>Intercept: {intercept:.2f}<br>R-squared: {r_value**2:.2f}"

# Add summary statistics annotation to the plot
fig.add_annotation(
    x=0.8, y=0.1,
    xref="paper", yref="paper",
    text=summary_text,
    showarrow=False,
    font=dict(size=12)
)

# Show the plot
fig.show()
```

```python
def lingress_misra_dropout(model_data, category, model_name):    
    new_model = model_data[model_data['category'] == category]

    slope, intercept, r_value, p_value, std_err = linregress(new_model['misra_score'], new_model['dropout_avg'])

    # Calculate the regression line
    regression_line = [(slope * x) + intercept for x in new_model['misra_score']]

    graph_title  = (str(category + ': Misra Vs. Dropout Average Linear Regression (' + model_name + ')'))
    # Create the scatter plot with regression line
    fig = px.scatter(new_model, x='misra_score', y='dropout_avg', title= graph_title, trendline='ols')

    # Add the regression line to the plot
    fig.add_traces(go.Scatter(x=new_model['misra_score'], y=regression_line, mode='lines', name='Regression Line'))

    # Create summary statistics text
    summary_text = f"Slope: {slope:.2f}<br>Intercept: {intercept:.2f}<br>R-squared: {r_value**2:.2f}"

    # Add summary statistics annotation to the plot
    fig.add_annotation(
        x=0.8, y=0.1,
        xref="paper", yref="paper",
        text=summary_text,
        showarrow=False,
        font=dict(size=12)
    )

    # Show the plot
    fig.show()
```

```python
def lingress_dropout_comp(model_1_data, model_1_name, model_2_data, model_2_name, category):    
    new_model = model_1_data[model_1_data['category'] == category]
    new_model['model_2_avg'] = model_2_data[model_2_data['category'] == category]['dropout_avg']

    slope, intercept, r_value, p_value, std_err = linregress(new_model['dropout_avg'], new_model['model_2_avg'])

    # Calculate the regression line
    regression_line = [(slope * x) + intercept for x in new_model['dropout_avg']]

    graph_title  = category + ': Dropout average regression ' + model_1_name + ' vs ' + model_2_name
    # Create the scatter plot with regression line
    fig = px.scatter(new_model, x='dropout_avg', y='model_2_avg', title= graph_title, trendline='ols')

    # Add the regression line to the plot
    fig.add_traces(go.Scatter(x=new_model['dropout_avg'], y=regression_line, mode='lines', name='Regression Line'))
    

    # Create summary statistics text
    summary_text = f"Slope: {slope:.2f}<br>Intercept: {intercept:.2f}<br>R-squared: {r_value**2:.2f}"

    # Add summary statistics annotation to the plot
    fig.add_annotation(
        x=0.8, y=0.1,
        xref="paper", yref="paper",
        text=summary_text,
        showarrow=False,
        font=dict(size=12)
    )
    fig.update_layout(xaxis_title = model_1_name + ' dropout average', yaxis_title = model_2_name + ' dropout average')
    # Show the plot
    fig.show()
```

```python
def lingress_rating(model_data, category, model_name):    
    new_model = model_data[model_data['category_x'] == category]

    slope, intercept, r_value, p_value, std_err = linregress(new_model['rating'], new_model['dropout_avg'])

    # Calculate the regression line
    regression_line = [(slope * x) + intercept for x in new_model['rating']]

    graph_title  = (str(category + ': Rating Vs. Dropout Average Linear Regression (' + model_name + ')'))
    # Create the scatter plot with regression line
    fig = px.scatter(new_model, x='rating', y='dropout_avg', title= graph_title, trendline='ols')

    # Add the regression line to the plot
    fig.add_traces(go.Scatter(x=new_model['rating'], y=regression_line, mode='lines', name='Regression Line'))

    # Create summary statistics text
    summary_text = f"Slope: {slope:.2f}<br>Intercept: {intercept:.2f}<br>R-squared: {r_value**2:.2f}"

    # Add summary statistics annotation to the plot
    fig.add_annotation(
        x=0.8, y=0.1,
        xref="paper", yref="paper",
        text=summary_text,
        showarrow=False,
        font=dict(size=12)
    )

    # Show the plot
    fig.show()
```

```python
def lingress_rating_variance(model_data, category, model_name):    
    new_model = model_data[model_data['category_x'] == category]

    slope, intercept, r_value, p_value, std_err = linregress(new_model['rating'], new_model['variance pop'])

    # Calculate the regression line
    regression_line = [(slope * x) + intercept for x in new_model['rating']]

    graph_title  = (str(category + ': Variance (popLM ' + model_name + ') Vs. Rosch Rating'))
    # Create the scatter plot with regression line
    fig = px.scatter(new_model, x='rating', y='variance pop', title= graph_title, trendline='ols')

    # Add the regression line to the plot
    fig.add_traces(go.Scatter(x=new_model['rating'], y=regression_line, mode='lines', name='Regression Line'))

    # Create summary statistics text
    summary_text = f"Slope: {slope:.2f}<br>Intercept: {intercept:.2f}<br>R-squared: {r_value**2:.2f}"

    # Add summary statistics annotation to the plot
    fig.add_annotation(
        x=0.8, y=0.1,
        xref="paper", yref="paper",
        text=summary_text,
        showarrow=False,
        font=dict(size=12)
    )


    # Show the plot
    fig.show()
    return fig
```
