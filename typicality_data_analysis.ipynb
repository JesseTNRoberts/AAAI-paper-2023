```python
import pandas as pd
import plotly.graph_objects as go
import researchpy as rp
import numpy as np

models = ['bert-base-uncased', 'bert-large-uncased', 'distilbert-base-uncased', 'distilgpt2',
          'gpt2-medium', 'gpt2', 'openai-gpt', 'roberta-base', 'roberta-large']

# 3 through 6 are CLMs
model = models[0]
stimulus = 'bird.'

misra_data_bert = pd.read_csv(loc+'rosch1975_alternate/'+model+".csv")
pop_data_bert_50 = pd.read_csv(loc+'rosch1975_alternate_popLM_50/'+model+".csv")
rosch_ratings = pd.read_csv(loc+'rosch1975_ratings.csv').drop('category', axis=1)

misra_data_bert.head()
misra_data_bert['score (sum, mean, [list)'].head()

mergedDF_50 = pd.merge(misra_data_bert, pop_data_bert_50, on=['item', 'category'], how='left')
mergedDF_50 = pd.merge(mergedDF_50, rosch_ratings, on=['item'], how='left')

mergedDF_50 = mergedDF_50.drop(columns=['stimulus_y','rank_y', 'params_y', 'model_y'])
mergedDF_50['score (sum, mean, [list)_x'] = mergedDF_50['score (sum, mean, [list)_x'].apply(ast.literal_eval)
mergedDF_50['score (sum, mean, [list)_y'] = mergedDF_50['score (sum, mean, [list)_y'].apply(ast.literal_eval)

mergedDF_50.rename(columns={'score (sum, mean, [list)_x': 'misra score (sum, mean, [list])'}, inplace=True)
mergedDF_50.rename(columns={'score (sum, mean, [list)_y': 'popLM score (sum, mean, [list])'}, inplace=True)

print(mergedDF_50['popLM score (sum, mean, [list])'].head())
print([[t[0] for t in m] for m in mergedDF_50['popLM score (sum, mean, [list])']])

# for sum
#select_reduction = 0
# for mean
select_reduction = 1

mergedDF_50['population data'] = [[t[select_reduction] for t in m] for m in mergedDF_50['popLM score (sum, mean, [list])']]
mergedDF_50['misra_avg'] = [t[select_reduction] for t in mergedDF_50['misra score (sum, mean, [list])']]

boxplots = []
misra_avg_markers = []

df = mergedDF_50.copy()

# Iterate through each row of the DataFrame and create a boxplot for each row
for index, row in df.iterrows():
    if(row['stimulus_x'] == stimulus):
        preamble = row['preamble_x']
        population_data = row['population data']
        boxplot = go.Box(y=population_data, name=preamble)
        boxplots.append(boxplot)
        misra_avg = row['misra_avg']

        misra_avg_marker = go.Scatter(x=[preamble], y=[misra_avg], mode='markers',
                                  name='Misra Avg', marker=dict(color='red', symbol='star'))
        misra_avg_markers.append(misra_avg_marker)

# Create the figure with all the boxplots
fig = go.Figure(data=boxplots + misra_avg_markers)

# Update the layout to provide a title and labels for the axes
fig.update_layout(
    title="Bird log distribution (misra shown as star)",
    xaxis_title="Preamble",
    yaxis_title="Population Data",
)

# Show the interactive plot in Jupyter Notebook
fig.show()
```

```python
def load_pop_data(model_name):
    misra_data = pd.read_csv('rosch1975_alternate/'+model_name+'.csv')
    pop_data = pd.read_csv('rosch1975_alternate_popLM_50/'+model_name+'.csv')
    misra_data.head()
    misra_data['score (sum, mean, [list)'].head()
    pop_data['score (sum, mean, [list)'][0]

    mergedDF_50 = pd.merge(misra_data, pop_data, on=['item', 'category'], how='left')
    mergedDF_50 = mergedDF_50.drop(columns=['stimulus_y','rank_y', 'params_y', 'model_y'])
    mergedDF_50['score (sum, mean, [list)_x'] = mergedDF_50['score (sum, mean, [list)_x'].apply(ast.literal_eval)
    mergedDF_50['score (sum, mean, [list)_y'] = mergedDF_50['score (sum, mean, [list)_y'].apply(ast.literal_eval)

    mergedDF_50.rename(columns={'score (sum, mean, [list)_x': 'misra score (sum, mean, [list])'}, inplace=True)
    mergedDF_50.rename(columns={'score (sum, mean, [list)_y': 'popLM score (sum, mean, [list])'}, inplace=True)

    #print(mergedDF_50['popLM score (sum, mean, [list])'].head())
    #print(mergedDF_50['popLM score (sum, mean, [list])'][0])
    #print(len(mergedDF_50['popLM score (sum, mean, [list])'][0]))

    mergedDF_50['population data'] = [[t[0] for t in m] for m in mergedDF_50['popLM score (sum, mean, [list])']]

    mergedDF_50['misra_avg'] = [t[0] for t in mergedDF_50['misra score (sum, mean, [list])']]
    mergedDF_50['dropout_avg'] = mergedDF_50['population data'].apply(np.average)
    mergedDF_50['std_deviation pop'] = mergedDF_50['population data'].apply(np.std)
    return mergedDF_50
```

```python
def correlate_data(df):
    df_to_corr = df.drop(columns=df.columns.difference(['rank_x', 'misra_avg', 'dropout_avg', 'std_deviation pop']))
    return df_to_corr.corr(method='spearman')
```
