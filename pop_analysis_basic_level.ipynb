```python
# Assuming your CSV file is named "data.csv"
import pandas as pd
import tqdm

import pandas as pd
import plotly.graph_objects as go
import researchpy as rp
import numpy as np
import ast

from google.colab import drive
drive.mount('/content/drive')

loc = "/content/drive/MyDrive/Data/BasicLevel/"
```

```python
models = ['bert-base-uncased', 'bert-large-uncased', 'distilbert-base-uncased', 'distilgpt2',
          'gpt2-medium', 'gpt2', 'openai-gpt', 'roberta-base', 'roberta-large']

# 3 through 6 are CLMs
model = models[5]

super_data = pd.read_csv(loc+'Superordinate/'+model+".csv")
basic_data = pd.read_csv(loc+'Basic/'+model+".csv")


super_data['score (sum, mean, [list)'] = super_data['score (sum, mean, [list)'].apply(ast.literal_eval)
basic_data['score (sum, mean, [list)'] = basic_data['score (sum, mean, [list)'].apply(ast.literal_eval)

# for sum
#select_reduction = 0
# for mean
select_reduction = 1

super_data['population data'] = [[t[select_reduction] for t in m] for m in super_data['score (sum, mean, [list)']]
basic_data['population data'] = [[t[select_reduction] for t in m] for m in basic_data['score (sum, mean, [list)']]


['Predict']
['Priming']

grouped = super_data.groupby('Subordinate')
super_data = grouped.aggregate(lambda x: [item for item in x])

grouped = basic_data.groupby('Subordinate')
basic_data = grouped.aggregate(lambda x: [item for item in x])
```

```python
def cohend(d1, d2) -> float:
  # calculate the size of samples
  n1, n2 = len(d1), len(d2)

  # calculate the variance of the samples
  s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)

  # calculate the pooled standard deviation
  s = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))

  # calculate the means of the samples
  u1, u2 = np.mean(d1), np.mean(d2)

  # return the effect size
  return (u1 - u2) / s


for df in [super_data, basic_data]:

  df['mean'] = df.apply(
      lambda row: [np.mean(item) for item in row['population data']],
      axis=1)

  df['std'] = df.apply(
      lambda row: [np.std(item) for item in row['population data']],
      axis=1)

  df['cohend - wrong'] = df.apply(
      lambda row: cohend(row['population data'][1], row['population data'][0]),
      axis=1)

  df['cohend - right'] = df.apply(
      lambda row: cohend(row['population data'][1], row['population data'][2]),
      axis=1)

  df['change - wrong'] = df.apply(
      lambda row: np.subtract(row['population data'][1], row['population data'][0]),
      axis=1)

  df['change - right'] = df.apply(
      lambda row: np.subtract(row['population data'][1], row['population data'][2]),
      axis=1)



super_data
```

```python
# This captures the change in probability at the basic and at the superordinate levels
# This shows clearly that the change in probability is much greater with higher deviation at the superordinate level
# and that it is highly significant

print(np.mean(abs(super_data['cohend - right']) - abs(basic_data['cohend - right'])))
print(np.mean(abs(super_data['cohend - wrong']) - abs(basic_data['cohend - wrong'])))

print('--------')

print(cohend([abs(element) for item in super_data['change - right'] for element in item],
             [abs(element) for item in basic_data['change - right'] for element in item]))

this = pd.DataFrame(list(zip([element for item in super_data['change - wrong'] for element in item],
 [element for item in basic_data['change - wrong'] for element in item])), columns=['super right', 'basic right'])

this["id"] = range(1, this.shape[0] + 1)

df2 = pd.melt(this, id_vars = "id", value_vars = ["super right", "basic right"],
                  var_name = "superBasic", value_name = "change_in_prob")

print(rp.difference_test("change_in_prob ~ C(superBasic)",
                   data = df2,
                   equal_variances = True,
                   independent_samples = True).conduct(effect_size = "all"))


this = pd.DataFrame(list(zip([element for item in super_data['change - right'] for element in item],
 [element for item in basic_data['change - right'] for element in item])), columns=['super wrong', 'basic wrong'])

this["id"] = range(1, this.shape[0] + 1)

df2 = pd.melt(this, id_vars = "id", value_vars = ["super wrong", "basic wrong"],
                  var_name = "superBasic", value_name = "change_in_prob")

print(rp.difference_test("change_in_prob ~ C(superBasic)",
                   data = df2,
                   equal_variances = True,
                   independent_samples = True).conduct(effect_size = "all"))

```
